```python
# Importación de librerias
import pandas as pd
import numpy as nd
import matplotlib.pyplot as plt

#cargar contenido desde la URL
url="https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
df=pd.read_csv(url)

#Mostrar primeras 5 filas
print(df.head(5))

#Mostramos los tipos que hay en las columnas
print(df.columns)

#Verificamos si hay valores nulos
print(df.isnull().sum())

#Eliminar datos nulos de la columna horsepower
mediana_hp = df['horsepower'].median()
df['horsepower'] = df['horsepower'].fillna(mediana_hp)
print(df.isnull().sum())

#Agregar una nueva columna llamda "relacion_peso_potencia"
df['relacion_peso_potencia']=df['weight']/df['horsepower']
print(df.head(5))
```

```python
# PARTE 2
#Histograma: Distribución de la aceleración
plt.figure(figsize=(10, 6))
plt.hist(df['acceleration'], bins=15, color='red', edgecolor='black')
plt.title('Distribución de la Aceleración de los Vehículos')
plt.xlabel('Aceleración')
plt.show()

# Gráfico de Dispersión: Displacement vs Horsepower
plt.figure(figsize=(10, 6))
plt.scatter(df['displacement'], df['horsepower'], alpha=0.3, color='blue')
plt.title('Relación: Desplazamiento del Motor vs Caballos de Fuerza')
plt.xlabel('Displacement')
plt.ylabel('Horsepower')
plt.show()

# Gráfico de Barras: Promedio de mpg agrupado por origin
plt.figure(figsize=(10, 6))
promedio_mpg = df.groupby('origin')['mpg'].mean()
promedio_mpg.plot(kind='bar', color=['green', 'gold', 'orange'])
plt.title('Promedio de Rendimiento por país de origen')
plt.xlabel('País de Origen')
plt.ylabel('MPG')
plt.xticks(rotation=0)
plt.show()
```
